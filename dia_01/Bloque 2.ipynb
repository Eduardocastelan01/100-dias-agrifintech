```python
def calcular_ingreso_bruto(hectareas, rendimiento_ton_ha, precio_por_ton):
    produccion_total = hectareas * rendimiento_ton_ha
    ingreso = produccion_total * precio_por_ton
    return ingreso

ingreso_veracruz = calcular_ingreso_bruto(
    hectareas=20,
    rendimiento_ton_ha=69.7,
    precio_por_ton=841.70
)
print(f"Ingreso bruto estimado: ${ingreso_veracruz:,.0f} pesos")
```

```python
def calcular_rentabilidad(hectareas, rendimiento, precio=1090, costo_por_ha=18000):
    ingreso_bruto = hectareas * rendimiento * precio
    costos_totales = hectareas * costo_por_ha
    ingreso_neto = ingreso_bruto - costos_totales
    margen = (ingreso_neto / ingreso_bruto) * 100
    return {
        "ingreso_bruto": ingreso_bruto,
        "costos": costos_totales,
        "ingreso_neto": ingreso_neto,
        "margen_pct": round(margen, 1)
    }

resultado_ingenio = calcular_rentabilidad(20, 56.6, precio=960)
print("--- Escenario Ingenio ---")
for clave, valor in resultado_ingenio.items():
    print(f"  {clave}: {valor}")

resultado_coyote = calcular_rentabilidad(20, 56.6, precio=300)
print("\n--- Escenario Coyote ---")
for clave, valor in resultado_coyote.items():
    print(f"  {clave}: {valor}")
```

```python
locales = {
    "Local 1 - Zona A": 9500,
    "Local 2 - Zona A": 11000,
    "Local 3 - Zona A": 13000,
    "Local 4 - Zona A": 10500,
    "Local 5 - Zona A": 9000,
    "Local 6 - Zona A": 12000,
    "Espectacular": 3000,
    "Bodega 1 - Zona B": 9500,
    "Bodega/Local - Zona B": 11000,
}

def reporte_arrendamientos(locales_dict):
    total_mensual = sum(locales_dict.values())
    total_anual = total_mensual * 12
    promedio = total_mensual / len(locales_dict)
    
    print("=" * 45)
    print("  REPORTE DE ARRENDAMIENTOS — EMPRESA FAMILIAR")
    print("=" * 45)
    for local, renta in sorted(locales_dict.items(), 
                                key=lambda x: x[1], reverse=True):
        print(f"  {local:<25} ${renta:>8,}")
    print("-" * 45)
    print(f"  Total mensual:             ${total_mensual:>8,}")
    print(f"  Total anual estimado:      ${total_anual:>8,}")
    print(f"  Promedio por local:        ${promedio:>8,.0f}")
    print("=" * 45)


    return total_mensual, total_anual

mensual, anual = reporte_arrendamientos(locales)
```
