```python
locales = {
    "Local 1 - Zona A": 9500,
    "Local 2 - Zona A": 11000,
    "Local 3 - Zona A": 13000,
    "Local 4 - Zona A": 10500,
    "Local 5 - Zona A": 9000,
    "Local 6 - Zona A": 12000,
    "Espectacular":      3000,
    "Bodega 1 - Zona B": 9500,
    "Bodega/Local - Zona B": 11000,
}

def reporte_arrendamientos(locales_dict):
    total_mensual = sum(locales_dict.values())
    total_anual   = total_mensual * 12
    promedio      = total_mensual / len(locales_dict)
    
    print("=" * 48)
    print("  REPORTE DE ARRENDAMIENTOS — EMPRESA FAMILIAR")
    print("=" * 48)
    for local, renta in sorted(locales_dict.items(),
                               key=lambda x: x[1], reverse=True):
        print(f"  {local:<28} ${renta:>8,}")
    print("-" * 48)
    print(f"  Total mensual estimado:    ${total_mensual:>8,}")
    print(f"  Total anual estimado:      ${total_anual:>8,}")
    print(f"  Promedio por propiedad:    ${promedio:>8,.0f}")
    print("=" * 48)
    return total_mensual, total_anual

mensual, anual = reporte_arrendamientos(locales)
```
